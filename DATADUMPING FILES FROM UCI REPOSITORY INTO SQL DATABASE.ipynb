# ACCESSING THE CSV FILE AND DUMPING DATA INTO DATABASE TABLE-

```python
with open("carbon_nanotubes.csv","r")as f1:
    next(f1)
    num_data=csv.reader(f1,delimiter=";")
    for i in num_data:
        l=[]
        print(i)
        for j in i:
            if "," in j:
                l.append(j.split(",")[0])
                l.append(j.split(",")[1])
            else:
                l.append(j)
        o=[int(k) for k in l]
        cur.execute("insert into wedeve.m values({},{},{},{},{},{},{},{},{},{},{},{},{},{})".format(o[0],o[1],o[2],o[2],o[3],o[4],o[5],o[6],o[7],o[8],o[9],o[10],o[11],o[12],o[13]))
        conn.commit()
```
