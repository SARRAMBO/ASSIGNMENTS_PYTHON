```python
url1="https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

OBJECTIVE- In the above we are given a raw dataset of a hotel's orders of different items done by different customers.
Our aim here is to pull or scrap the data through pandas api and using various pandas operations,pick out key insights and arrange the data in neat tabular format(eligible for dumping in SQL database) which might help in increasing the business of the hotel by manifolds.Other data cleansing operations are also to be kepy in priority,like arranging columns in alphabetically sorted format,filtering out null values from certain columns and indexes etc.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import numpy
```

```python
df=pd.read_csv(url1,sep='\t')
df #now pandas data frame object is created which converts raw data into neat,tabular format.
```

```python
#printing names of all columns
df.columns
df.dtypes.index
#How pandas is indexing the dataset-
df.index
```

# MOST ORDERED ITEM AND HOW MANY NO OF TIMES IT WAS ORDERED-

```python
c=df.groupby('item_name').sum() #grouping all columns by 'item_name' and adding all similar item_names in terms of numerical columns(qunatity,order_id) 
c=c.sort_values(['quantity'],ascending=False)#sorting in terms of numerical column 'quantity' in descending order.
c.head(1)#Most ordered item
c.tail(1)#least ordered item
c.head(1)
```

# MOST ORDERED ITEM IN CHOICE DESCRIPTION COLUMN-

```python
c=df.groupby('choice_description').sum()
c=c.sort_values(['quantity'],ascending=False)
c.head(1) #most ordered
c.tail(1) #least ordered item
```

# ITEMS HAVING NO CHOICE DESCRIPTION AND MOSTLY ORDERED-

```python
df1=df.iloc[df['choice_description'][df['choice_description'].isnull()==True].index]
#filtering out null values in the choice description column and storing it in another dataframe object.
df1=df[df.choice_description.isnull()]
```

```python
c=df1.groupby('item_name').sum()
c=c.sort_values(['quantity'],ascending=False)
c.head(1) #Most ordered
c.tail(1) #least ordered.
```

```python
#converting column 'item_price' data type into numerical data-
remove_dollar=lambda x:float(x[1:-1])
df["item_price"]=df['item_price'].apply(remove_dollar)
```

```python
df #now $ removed from item_price and converted to float datatype.
```

# ITEM GENERATING HIGHEST REVENUE-

```python
c=df.groupby('item_name').sum()
c=c.sort_values(['item_price'],ascending=False)
c.head(1) #highest revenue generating item
c.tail(1) #lowest revenue generating item
```

# CONTRIBUTION OF ITEMS TOWARDS TOTAL REVENUE-

```python
df["revenue"]=df["item_price"]*df["quantity"]
df3=df.groupby('item_name').sum()
df3=df3.sort_values('revenue',ascending=False)
sw=df3.head(1)['revenue']["Chicken Bowl"]
sm=df.revenue.sum()
sw/sm #contribution of highest ordered and highest sold item as a percentage of total revenue
print("Contribution of Chicken Bowl towards the total revenue is:",np.round((sw/sm),2))
ss=df3.tail(1)['revenue']['Chips and Mild Fresh Tomato Salsa']
print("Contribution of Chips and Mild Fresh Tomato Salsa towards total revenue is:",np.round((ss/sm),5))
#contribution of lowest sold item as a percentage of total revenue


```

# ITEM IN CHOICE_DESCRIPTION FETCHING HIGHEST REVENUE-

```python
c=df.groupby('choice_description').sum()
c=c.sort_values(['item_price'],ascending=False)
c.head(1) #item fetching most revenue in choice description column
c.tail(1) #item fetching least revenue in choice description column
```

# REVENUE GENERATION FOR THE GIVEN PERIOD-

```python
revenue=(df["quantity"]*df['item_price']).sum()
print("TOTAL REVENUE GENERATION FOR THE GIVEN PERIOD IS $:",str(np.round(revenue,2)))
df.revenue.sum()
```

# DELETING DUPLICATES IN ITEM_NAME AND QUANTITY-

```python
filtered=df.drop_duplicates(['item_name','quantity'])
filtered
```
